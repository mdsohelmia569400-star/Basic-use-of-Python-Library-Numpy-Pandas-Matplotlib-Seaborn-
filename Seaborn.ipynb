```python
#####.....Seaborn line plot..........##########
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset('titanic').head(20)



sns.lineplot(x='age',y = 'fare',data = df, hue = 'sex',style='sex', palette = 'Accent',markers = ['*','D'],markersize = 15,legend = True)
plt.grid()
plt.title('Titanic Dataset')









plt.show()
```

```python
df = pd.read_excel('for.xlsx')

sns.lineplot(x = 'id',y = 'age',data = df, palette = 'Accent',marker = '*',markersize = 15,
             legend = True,hue = 'bonous')

plt.grid()
plt.title('this is seaborn plot method ')






plt.show()
```

```python
#######.........Seaborn Bar Plot...........########

df = sns.load_dataset('titanic').head(30)
df
```

```python
sns.set(style = 'darkgrid')
sns.set_palette('pastel6')
sns.barplot(x = 'class', y = 'fare',data = df, palette = 'pastel6',hue = 'sex',orient = 'v', ci = 100 , n_boot = 3,saturation = 10)
plt.grid()
```

```python
#####........Histogram plot in seaborn..........########

var = sns.load_dataset('penguins')
var.head(10)


```

```python
sns.displot(var['bill_length_mm'],bins=[35,40,45,50,55,60],kde = True,rug = True,color = 'red')
plt.show()
```

```python
sns.displot(var['bill_length_mm'],kde = True,rug = True,color = 'red',log_scale = True)
plt.show()
```

```python
#####......... Seaborn Sceatter plot........########


all = sns.load_dataset('penguins')
all.head(10)
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = all,hue = 'sex', style = 'sex',size = 'sex', sizes = (90,90),
                palette = 'Accent',alpha = 1,legend = True, markers = ['*','D'])
plt.grid()
plt.show()
```

```python
#####...........Heatmap Plot in Seaborn...........##########

import numpy as np


var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
y = {'fontsize' : 20,'color': 'red'}
sns.heatmap(var,annot = True, annot_kws=y,linewidths = 10, linecolor = 'y')
```

```python
data = sns.load_dataset('anagrams')
x = data.drop(columns = ['attnr'], axis = 1).head(10)
x
```

```python
sns.heatmap(x,vmin = 0, vmax = 7,cmap = 'Accent')
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
data2 = sns.load_dataset('anagrams').head(20)
data2

pivot_data = data2.pivot(index = 'subidr', columns = 'attnr', values = 'num1' )
sns.heatmap(pivot_data,annot = True, cmap = 'Accent')
plt.show()
```

```python
#######..........Count Plot in Seaborn.........##########


var3 = sns.load_dataset('tips')
var3.head(10)
```

```python
sns.countplot(x = 'sex', data = var3, palette = 'Accent', hue = 'smoker')
plt.show()
```

```python
#####........Vilon plot in seaborn.........##########

var5 = sns.load_dataset('tips')
var5
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = var5, palette = 'PuOr', hue = 'sex', linewidth = 1, saturation = .4 )
plt.show()
```

```python
sns.violinplot(x = 'time', y = 'total_bill', data = var5, palette = 'PuOr', order = ['Dinner','Lunch'])
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = var5, palette = 'PuOr', hue = 'sex', linewidth = 1, saturation = .4 , split = True)
plt.show()
```

```python
sns.violinplot(y = var5['total_bill'], color = 'red')
plt.show()        #### It's a single violinplot
```

```python
########........Pairplot in seaborn..........##########

var6 = sns.load_dataset('tips')
var6
```

```python
sns.pairplot(var6, hue = 'sex', palette = 'Accent')
plt.show()
```

```python
sns.pairplot(var6,vars = ['total_bill','tip'], hue = 'sex', palette = 'Accent', hue_order = ['Female','Male'], kind = 'reg')
plt.show()                         ###### kind = scatter, kde, hist, reg
```

```python
########..........Strip plot in seaborn............###############
```

```python
var7 = sns.load_dataset('tips')
var7
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = var7, hue = 'sex', palette = 'Spectral', jitter = True, dodge = True, linewidth = 1,
              edgecolor = 'r',size = 10, alpha = .6)
plt.show()
```

```python
sns.stripplot(y = var7['total_bill'])
```

```python
###########........Boxplot in Seaborn............###########
var8 = sns.load_dataset('tips')
var8
```

```python
sns.set(style = 'whitegrid')
sns.boxplot(x = 'day', y = 'total_bill', data = var8, palette = 'Accent', hue = 'sex', linewidth = 1,showmeans = True)


plt.show()
```

```python
sns.set(style = 'whitegrid')
sns.boxplot(y = var8['total_bill'], color = 'red')
plt.show()
```

```python
#######.........Factor plot/catplot in seaborn...........#######

var9 = sns.load_dataset('tips')
var9
```

```python
sns.catplot(x = 'size', y = 'tip', data = var9, hue = 'sex', palette = 'Accent',kind = 'point', height = 5)
                                       ######## kind = point, bar, count, violin, box, boxen, strip, swarm
plt.show()
```

```python
#######...........multiple plot (Facet-Grid) in seaborn..........##########

var10 = sns.load_dataset('tips')
var10
```
